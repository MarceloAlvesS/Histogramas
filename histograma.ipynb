```python
tempo = pd.read_csv(url)['Tempo (s)'].map(lambda x: x.replace(',', '.')).to_numpy(dtype=float)

n = len(tempo)
faixa = max(tempo) - min(tempo)
classes = np.ceil(1 + 3.3*np.log10(n))
contagens, limites = np.histogram(tempo, bins=int(classes))
centros = (limites[:-1] + limites[1:]) / 2

larguras = np.diff(limites)

# Plotando com as bordas exatas
plt.hist(tempo, bins=limites, edgecolor='black', color='#8B0000')

bin_width = np.diff(limites)[0]
if bin_width < 0.5:
    plt.xticks(rotation=45, fontsize=8)
    plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(round(bin_width * 5, 2)))
else:
    plt.xticks(rotation=0, fontsize=10)
    plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(round(bin_width, 2)))

# Iterando pelas barras do histograma
for i in range(len(contagens)):
    valor_central = centros[i]
    altura_barra = contagens[i]

    # Adiciona o ponto no meio da barra
    plt.plot(valor_central, altura_barra, 'ko', markersize=3)  # 'ko' para ponto preto

    # Adiciona o texto com o valor central acima do ponto
    plt.text(valor_central, altura_barra+0.2, f'{valor_central:.3f}',
             ha='center', va='bottom', fontsize=10, color='black')

plt.xlabel('Tempo (s)')
plt.ylabel('Frequência')
plt.title('Histograma do tempo Marcelo x Débora')
plt.xticks(limites)  # coloca os ticks exatamente nas bordas

plt.show()
```
