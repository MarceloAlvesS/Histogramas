```python
tempo = pd.read_csv(url)['Tempo (s)'].map(lambda x: x.replace(',', '.')).to_numpy(dtype=float)

plt.figure(figsize=(6, 4))
plt.scatter(range(len(tempo)), tempo,
            s=10,      # tamanho do ponto
            alpha=0.7, # transparência
            color='#af0000')

plt.axhline(0.4517, color='black', linewidth=2, label='y = 9.8', alpha=.7)

plt.ylim(0.2, 2)  # <- fixa os limites do eixo Y

plt.ylabel('Tempo s')
plt.title('Distribuição dos tempos')
plt.grid(True, axis='y', linestyle='', alpha=0.5)
plt.legend(['Pontos', 'Previsão'])
plt.show()
```

```python
aceleracao = pd.read_csv(url)['g (m/s²)'].map(lambda x: x.replace(',', '.')).to_numpy(dtype=float)

plt.figure(figsize=(6, 4))
plt.scatter(range(len(aceleracao)), aceleracao,
            s=10,      # tamanho do ponto
            alpha=0.7, # transparência
            color='#008B00')

plt.axhline(9.8, color='black', linewidth=2, label='y = 9.8', alpha=.7)

plt.ylim(0, 25)  # <- fixa os limites do eixo Y

plt.ylabel('Aceleração (m/s²)')
plt.title('Distribuição das Gravidades')
plt.grid(True, axis='y', linestyle='', alpha=0.5)
plt.legend(['Pontos', 'Previsão'])
plt.show()
```
